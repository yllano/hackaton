```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
file_path = r"c:\Users\anaca\OneDrive\Escritorio\hackaton\01_data\processed\OnlineRetail_clean.csv"

# Leer solo primeras filas para inspecci√≥n inicial
df= pd.read_csv(file_path, encoding='ISO-8859-1', nrows=1000)
print("=== Primeras filas ===")
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['UnitPrice'], bins=50)
plt.xlabel('Precio Unitario')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n del Precio Unitario (UnitPrice)')
plt.grid(True)
plt.show()
plt.figure(figsize=(10,6))
plt.hist(df['Quantity'], bins=50)
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de la Cantidad (Quantity)')
plt.grid(True)
plt.show()
```

Muchos precios bajos

Pocos precios extremadamente altos

```python
import matplotlib.pyplot as plt

sales_by_hour = df.groupby('InvoiceHour')['Quantity'].sum()

plt.figure(figsize=(10,5))
plt.bar(sales_by_hour.index, sales_by_hour.values)
plt.title("Distribuci√≥n de ventas por hora")
plt.xlabel("Hora")
plt.ylabel("Cantidad vendida")
plt.xticks(range(7, 21))
plt.grid(True)
plt.show()

```

```python
sales_by_minute = df.groupby('InvoiceMinute')['Quantity'].sum()

plt.figure(figsize=(12,5))
plt.plot(sales_by_minute.index, sales_by_minute.values, marker='o')
plt.title("Distribuci√≥n de ventas por minuto del d√≠a")
plt.xlabel("Minuto (0-59)")
plt.ylabel("Cantidad vendida")
plt.grid(True)
plt.show()
```
